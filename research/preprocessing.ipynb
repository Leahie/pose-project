```python
import pandas as pd 
df = pd.read_json("../data/mpii_annotations.json")
```

```python
df_filtered = df[df['numOtherPeople']==0]
```

```python
# Shuffle deterministically, keep first 1,000 rows, and save as sample data
shuffled_df = df_filtered.sample(frac=1, random_state=42).reset_index(drop=True)
sample_data = shuffled_df.copy()

sample_data.to_csv("../data/data.csv", index=False)
sample_data.head()
```
